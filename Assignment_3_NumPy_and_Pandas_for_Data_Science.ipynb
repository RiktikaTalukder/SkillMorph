# **Task 1.2: NumPy Array Operations**

```python
# Convert score columns to numpy arrays
math_scores = df['math score'].values
reading_scores = df['reading score'].values
writing_scores = df['writing score'].values

# 1. Calculate mean, median, std, min, max for each subject using NumPy
columns = ['Math', 'Reading', 'Writing']
for column, score in zip(columns, [math_scores, reading_scores, writing_scores]):
  mean = np.mean(score)
  median = np.median(score)
  std = np.std(score)
  min = np.min(score)
  max = np.max(score)
  print(f"{column} - Mean:",mean,", Median:",median,", Std:",std,", Min:",min,", Max:",max)

# 2. Find total students who scored above 80 in math
print("\nStudents with math > 80:",sum(math_scores>80))

# 3. Find total students who scored below 50 in any subject
print("\nStudents with any score < 50:",sum((math_scores<50) | (reading_scores<50) | (writing_scores<50)))

# 4. Calculate the overall average score across all three subjects
overall_average = np.mean([math_scores, reading_scores, writing_scores])
print("\nOverall Average Score:",overall_average)
```

```python
# Artificially create missing data for practice
# 1. Create copy and show original missing values
df_practice = df.copy()
print("Original missing values:")
print(df_practice.isnull().sum())

np.random.seed(42)  # For reproducible results

# 2. Randomly set 50 math scores to np.nan
missing_math_idx = np.random.choice(df_practice.index, 50, replace=False)
df_practice.loc[missing_math_idx, 'math score'] = np.nan
print("\nmissing value in math score:",df_practice['math score'].isnull().sum())

# 3. Randomly set 30 writing scores to np.nan
missing_writing_idx = np.random.choice(df_practice.index, 30, replace=False)
df_practice.loc[missing_writing_idx, 'writing score'] = np.nan
print("missing value in writing score:",df_practice['writing score'].isnull().sum())

# 4. Fill missing math scores with the MEDIAN math score
median = (df_practice['math score']).median()
df_practice['math score'] = df_practice['math score'].fillna(median)

# 5. Fill missing writing scores with the MEAN writing score
mean = (df_practice['writing score']).mean()
df_practice['wrting score'] = df_practice['writing score'].fillna(mean)

# 6. Verify no missing values remain
print("\n",df_practice.isnull().sum())
```

# **Task 3.2: Basic Groupby Analysis**

```python
# 1. Group by 'gender' and calculate:
#    - Average scores in all subjects
gender_group = df.groupby('gender')
print('Average scores by gender:')
score = ['math score','reading score','writing score']
display(gender_group[score].mean())
#    - Count of students
print("\nCount of students by gender:")
display(gender_group['gender'].count())

# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
print("\nMean scores for each subject")
display(df.groupby('test preparation course')[score].mean())
#    - Show which subject benefits most from test prep
print("\nSubject that benefits most from test prep")
display(df.groupby('test preparation course')[score].mean().max())

# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
print("\nAverage math score for each education level")
display(df.groupby('parental level of education')['math score'].mean())
#    - Rank education levels by math performance
print("\nRank education levels by math performance")
display(df.groupby('parental level of education')['math score'].mean().sort_values(ascending=False))

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
print("\nAverage scores for each subject")
display(df.groupby('lunch')[score].mean())
#    - Count of students in each lunch category
print("\nNumber of students in each lunch category")
display(df.groupby('lunch')['lunch'].count())
```
